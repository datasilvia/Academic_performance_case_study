```python
# Load data
df = pd.read_csv('df_limpito_EDA.csv')
```

```python
# Establecer el estilo de Seaborn
sns.set(style="whitegrid")

# DistribuciÃ³n de variables numÃ©ricas
df.hist(bins=30, figsize=(20, 15), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Variables', fontsize=20)
plt.show()
```

```python
# DistribuciÃ³n de las variables categÃ³ricas
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(12, 6))
    # Countplot
    sns.countplot(data=df, x=column, palette='viridis')
    plt.title(f'Countplot of {column}', fontsize=18)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.xticks(rotation=45, fontsize=12)
    plt.yticks(fontsize=12)
    sns.despine()
    plt.show()

    # Pie chart
    plt.figure(figsize=(8, 8))
    df[column].value_counts().plot.pie(autopct='%1.1f%%', colors=sns.color_palette('viridis', len(df[column].unique())), startangle=90, wedgeprops={'edgecolor': 'black'})
    plt.title(f'Pie Chart of {column}', fontsize=18)
    plt.ylabel('')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
```
