```python
# Import librerias for EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Load data
df = pd.read_csv('df_limpito_EDA.csv')
```

```python
# Show data
df.head()
```

```python
# Show data types
df.dtypes
```

```python
# Crear nuevas columnas 'por_status' y 'math_status' basadas en 'por_score' y 'math_score'
df['por_status'] = df['por_score'].apply(lambda x: 'aprobado' if x >= 10 else 'suspenso')
df['math_status'] = df['math_score'].apply(lambda x: 'aprobado' if x >= 10 else 'suspenso')

# Mostrar las primeras filas para verificar
df[['por_score', 'por_status', 'math_score', 'math_status']].head()
```

```python
# Info data
df.info()
```

```python
# Describe data
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Establecer el estilo de Seaborn
sns.set_style('darkgrid')

# DistribuciÃ³n de variables numÃ©ricas
df.hist(bins=30, figsize=(20, 15), color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numerical Variables', fontsize=20)
plt.show()
```

```python
# 1. Age Distribution

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='age', palette='viridis')
plt.title('Bar Plot of Age Distribution', fontsize=18)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
sns.despine()
plt.show()
```

```python
# 2.Gender Distribution 

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='sex', palette='viridis')
plt.title('Distribution of Gender', fontsize=18, fontweight='bold')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
sns.despine()
plt.show()
```

```python
# Contar la cantidad de cada gÃ©nero
gender_counts = df['sex'].value_counts()

# Crear el grÃ¡fico de pastel
plt.figure(figsize=(8, 8))
colors = sns.color_palette('viridis', len(gender_counts))
explode = (0.05, 0.05)  # SeparaciÃ³n entre las porciones

plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, shadow=True, wedgeprops={'edgecolor': 'black'})
plt.title('Distribution of Gender', fontsize=18, fontweight='bold')
plt.axis('equal')  # Para asegurar que el grÃ¡fico sea un cÃ­rculo
plt.show()
```

```python
# 3.Study time by Gender

plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='studytime', hue='sex', kde=True, palette='viridis', multiple='stack')
plt.title('Distribution of Study Time by Gender', fontsize=15)
plt.xlabel('Study Time')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analysis of the dataset

# Distribution of mother's job
mjob_distribution = df['Mjob'].value_counts(normalize=True) * 100

# Distribution of father's job
fjob_distribution = df['Fjob'].value_counts(normalize=True) * 100

# Distribution of romantic status
romantic_distribution = df['romantic'].value_counts(normalize=True) * 100

# Average scores in Portuguese and Math by gender
avg_scores_by_gender = df.groupby('sex')[['por_score', 'math_score']].mean()

# Average scores in Portuguese and Math by romantic status
avg_scores_by_romantic = df.groupby('romantic')[['por_score', 'math_score']].mean()

# Average scores in Portuguese and Math by mother's job
avg_scores_by_mjob = df.groupby('Mjob')[['por_score', 'math_score']].mean()

# Average scores in Portuguese and Math by father's job
avg_scores_by_fjob = df.groupby('Fjob')[['por_score', 'math_score']].mean()

# Distribution of absences in Portuguese
absences_por_distribution = df['absences_por'].describe()

# Distribution of absences in Math
absences_math_distribution = df['absences_math'].describe()

# Distribution of travel time
traveltime_distribution = df['traveltime'].describe()

# Distribution of free time
freetime_distribution = df['freetime'].describe()

# Generate the analysis in Markdown format
markdown_analysis = f"""
# Statistical Analysis of the Dataset

## Distribution of Mother's Job
{mjob_distribution.to_markdown()}

## Distribution of Father's Job
{fjob_distribution.to_markdown()}

## Distribution of Romantic Status
{romantic_distribution.to_markdown()}

## Average Scores in Portuguese and Math by Gender
{avg_scores_by_gender.to_markdown()}

## Average Scores in Portuguese and Math by Romantic Status
{avg_scores_by_romantic.to_markdown()}

## Average Scores in Portuguese and Math by Mother's Job
{avg_scores_by_mjob.to_markdown()}

## Average Scores in Portuguese and Math by Father's Job
{avg_scores_by_fjob.to_markdown()}

## Distribution of Absences in Portuguese
{absences_por_distribution.to_markdown()}

## Distribution of Absences in Math
{absences_math_distribution.to_markdown()}

## Distribution of Travel Time
{traveltime_distribution.to_markdown()}

## Distribution of Free Time
{freetime_distribution.to_markdown()}
"""

# Save the analysis to a Markdown file
with open('analysis.md', 'w') as f:
    f.write(markdown_analysis)

print("Analysis saved to 'analysis.md'")
```
